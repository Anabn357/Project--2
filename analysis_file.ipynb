```python
# import libraries

import csv
import pandas as pd
```

```python
# source 1: kaggale.com

url_1 = "rating_sale_info.csv"

kaggle_data = pd.read_csv(url_1)

kaggle_data
```

```python
# source 2: scraping whatoplay: "https://whatoplay.com/best?sort-by=gamer_score"

url_2 = "scraping_result.csv"

whatoplay_data = pd.read_csv(url_2)

whatoplay_data
```

```python
total_data = pd.merge(kaggle_data, whatoplay_data, how = "left", on=["Name", "Platform"] )

total_data
```

```python
# Store the final table as csv file

total_data.to_csv("data/csv/master_table.csv", index= False, header = True)
```

```python
# convert csv to jason format

total_data.to_json ("data/json/master_table.json")
kaggle_data.to_json ("data/json/kaggle_data.json")
whatoplay_data.to_json ("data/json/whatoplay_data.json")
```
